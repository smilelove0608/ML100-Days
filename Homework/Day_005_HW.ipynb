```python
#柱狀圖
fig, axes = plt.subplots(2, 1)
barpho = pd.Series(app_train['AMT_INCOME_TOTAL'], index=np.arange(0, 100, 10))
barpho.plot(kind='bar', ax=axes[0], color='k', alpha=0.7)
barpho.plot(kind='barh', ax=axes[1], color='b', alpha=0.7)
```

```python
#直方圖
app_train['AMT_INCOME_TOTAL'][0:10].hist()
```

```python
#密度圖(KDE圖)
app_train['AMT_INCOME_TOTAL'][0:10].plot(kind='kde')
```

```python
#直方圖與密度圖
#https://blog.csdn.net/huanhuan_Coder/article/details/80762208
comp1 = np.random.normal(0, 1, size=200)
comp2 = np.random.normal(10, 2, size=200)
values = pd.Series(np.concatenate([comp1, comp2]))
values.hist(bins=100, alpha=0.3, color='k', density=True)
values.plot(kind='kde', style='b')
```

```python
#散布圖
plt.scatter(app_train['AMT_INCOME_TOTAL'][0:100], app_train['AMT_ANNUITY'][0:100])
plt.title('100%s vs.100%s' % ('AMT_INCOME_TOTAL', 'AMT_ANNUITY'))
```
