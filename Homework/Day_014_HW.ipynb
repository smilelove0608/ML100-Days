### 請根據不同的 HOUSETYPE_MODE 對 AMT_CREDIT 繪製 Histogram

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = 'C:\\Users\\mara0608\\Desktop\\AI_Learning\\Open_living\\ML100DAYS_ming\\DAY2'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
app_train['HOUSETYPE_MODE'] #有NaN
#https://blog.csdn.net/brucewong0516/article/details/80406564
#https://blog.csdn.net/liulina603/article/details/78781893
app_train['HOUSETYPE_MODE'].fillna("Na",inplace=True)
app_train['HOUSETYPE_MODE'] #以Na取代
```

```python
app_train['HOUSETYPE_MODE'].unique()
np.sort(app_train['HOUSETYPE_MODE'].unique())
```

```python
len(np.sort(app_train['HOUSETYPE_MODE'].unique()))
```

```python
"""
Your Code Here
"""
unique_house_type = np.sort(app_train['HOUSETYPE_MODE'].unique())

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    app_train.loc[app_train['HOUSETYPE_MODE']==unique_house_type[i],'AMT_CREDIT' ].hist()
    plt.title(str(unique_house_type[i]))
plt.show()    
```
